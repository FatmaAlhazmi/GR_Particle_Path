```python
# Import necessary modules
import sys
import os
sys.path.append(os.path.abspath('..'))

from gr_particle_path_tool.metrics import schwarzschild_metric
from gr_particle_path_tool.utils import christoffel_symbols, geodesic_equations, solve_geodesic
from gr_particle_path_tool.visualizations import plot_geodesic
import sympy as sp
import numpy as np
from IPython.display import display

print("Modules imported")
```

```python
# Define the Schwarzschild metric
metric = schwarzschild_metric()
coords = sp.symbols('t r theta phi')

# Print metric to verify it is correctly defined
print("Metric defined")
display(metric)
```

```python
# Calculate Christoffel symbols
christoffel = christoffel_symbols(metric, coords)

# Print and display the Christoffel symbols
print("\nChristoffel Symbols:")
display(christoffel)
print(christoffel)
```

```python
# Define initial conditions and solve geodesic equations
initial_conditions = [0, 10, np.pi/2, 0, 0, -0.1, 0, 0.05]  # Example values
t_span = (0, 100)
t_eval = np.linspace(0, 100, 1000)
geodesic_func = geodesic_equations(christoffel, coords)
solution = solve_geodesic(geodesic_func, initial_conditions, t_span, t_eval)

# Plot the geodesic path
print("\nGeodesic Path Plot:")
plot_geodesic(solution, coords)
```
